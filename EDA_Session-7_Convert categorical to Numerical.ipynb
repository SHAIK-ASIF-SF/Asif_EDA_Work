**one hot encoder**

- one hot encoder name says at a time one will on and other will be off

- For example case_status has two labels
    - certified
    - Denied
    
- when you apply only hot encoding on case status, It creates two more extra columns

    - case_status_Certified
    - case_status_Denied


case_status case_status_certified case_status_denied

Certified           1                  0
Denied              0                  1

**Advantages**

- when you ML model it is very important the column should be independent each other

- so here case_status creating two extra columns

- Which are independent each other which means the row values at a time only one column has 1

- columns are independent each other

- Which means 90 degrees phase shift

- Which means orthoganal each other

**Dis-advantages**

- The Dis advantage is is a column has 100 unique lables, 100 new      columns will be created
- the data will become saparse, which means huge
- Columns are more means, Dimesnions are more
- The processing time is more
- The memory consumption is more

- Course of Dimensionality
