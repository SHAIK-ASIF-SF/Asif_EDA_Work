#### creat dataframes using list

```python
name=['Ramesh','Suresh','Sathish']
age=[30,35,40]
name,age
```

**Step-1**

$Create$ $data$ $frame$

```python
pd.DataFrame()  # make the data frame
```

**Step-2**

$Provide$ $data$

```python
pd.DataFrame(zip(name,age))
```

**Step-3**

$Provide$ $column$

```python
#Provide columns
data=zip(name,age)
cols=['Name','Age']
pd.DataFrame(data,columns=cols)

#pd.DataFrame(zip(name,age),columns=['Name','Age'])
```

**Step-4**

$Provide$ $index$

```python
data=zip(name,age)
cols=['Name','Age']
ind=['A','B','C']
df=pd.DataFrame(data,
            columns=cols,
            index=ind)
df
```

**Step-5**

Add new columns

```python
name=['Ramesh','Suresh','Sathish']
age=[30,35,40]

data=zip(name,age)
cols=['Name','Age']
ind=['A','B','c']
df=pd.DataFrame(data,columns=cols,index=ind)
df
```

- if ypu want to add a new column

- df['new column']

- you need to have a list which is having some elements

- that elements need to equal to number of rows

- city_names=['Hyd','Blr','chennai']

- df['city']=city_names

```python
city_names=['Hyd','Blr','Chennai']
df['city']=city_names
df
```

**Step-6**

$update$ $the$  $exsisting$ $column$

- if you want to create new column or update the old column

- both are same way

```python
df['Name']=['Asif','Omkar','Rocky']
df
```

**Step-7**

$drop$ $the$ $column$

- In order to drop the column

- we neeet to use drop methoud

- it takes 3 parameters

     - drop column or row
      
     - mention the column name
    
     - axis
          
       - axis=1 referance as column
              
       - axis=0 referance as row
                
    - you want to  overwrite the existing dataframe
    
    - you want overwrite the existing dataframe
         
       - inplace=True

```python
df.drop('city',   # column name
          axis=1, # column
         inplace=True)  # overwrite the same 
```

```python
df
```

```python
name=['Ramesh','Suresh','Sathish']
age=[30,35,40]


df=pd.DataFrame(zip(name,age),
               columns=['Name','Age'],
               index=['A','B','C'])

city_names=['Hyd','Blr','Chennai']
df['city']=city_names

df.drop('city', # column name
        axis=1,  # Column
       inplace=True)  # Overwrite the same
df



```

**Step-8**

$Drop$ $rows$

```python
df.drop('C',  # column name
        axis=0,   # Column
        inplace=True) # Overwrite the same
df
        
```

**Step-9**

$save$ $the$ $data$ $frame$

```python
df.to_csv("output.csv")
# while saving index consider as extra column
df.to_excel("output.xlsx")
```

```python
# read output csv
pd.read_csv("output.csv")
```

**Step-10**


$Remove$ $the$ $index$



```python
# To avoid the above problem
# give index=False
df.to_csv("output.csv",index=False)
```
