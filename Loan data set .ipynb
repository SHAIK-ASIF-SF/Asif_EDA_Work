```python
unique_lables=loan_df['Self_Employed'].unique()
count=[]
for i in unique_lables:
    con=loan_df['Self_Employed']==i
    count.append(len(loan_df[con]))

Self_Employed_df=pd.DataFrame(zip(unique_lables,count),
                              columns=['Self_Employed','count'])
Self_Employed_df.to_csv('Self_Employed.csv',index=False)
```
